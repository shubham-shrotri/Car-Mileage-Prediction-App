```python
from pycaret.regression import *
from pycaret.datasets import get_data
```

```python
import pandas as pd
import numpy as np
data = pd.read_csv('auto_mpg.csv')
```

```python
data.info()
```

```python
data.drop(['model_year',''])
```

```python
data.drop(data.loc[data['horsepower'] == '?'].index,inplace = True)
```

```python
data.drop(['model year','car name'],axis = 1, inplace = True)
```

```python
r2 = setup(data, target = 'mpg', session_id = 123,
           numeric_features = ['cylinders','horsepower'],
           normalize = True,
           polynomial_features = True, trigonometry_features = True,
           feature_interaction=True)
```

```python
lr = create_model('lr')
```

```python
save_model(lr, 'deployment_2_07052020')
```
